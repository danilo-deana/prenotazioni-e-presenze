```python
import pandas as pd

dfHRS = pd.read_excel('xlsx' + '/' + 'orari.xlsx')

dfHRS.info()
```

```python
def giorno_della_settimana(giorno):
    if giorno == "L":
        return 0
    elif giorno == "Ma":
        return 1
    elif giorno == "Me":
        return 2
    elif giorno == "G":
        return 3
    elif giorno == "V":
        return 4
    elif giorno == "S":
        return 5
    elif giorno == "D":
        return 6

def giornate(giorni):
    giornate = 0
    for intervallo in giorni.split(';'):
        if intervallo.find("-") == -1:
            giornate += 1
        else:
            giornate += (giorno_della_settimana(intervallo[intervallo.find("-") + 1:]) - giorno_della_settimana(intervallo[0:(intervallo.find("-"))])) + 1
    return giornate
            
dfHRS['giornate'] = dfHRS.apply(lambda x: giornate(x['giorni']), axis = 1)
```

```python
def giorni_della_settimana(giorni):
    giorni_della_settimana = ""
    for intervallo in giorni.split(';'):
        if intervallo.find("-") == -1:
            if giorni_della_settimana == "":
                giorni_della_settimana = intervallo
            else:
                giorni_della_settimana = giorni_della_settimana + ";" + intervallo
        else:
            giorno_iniziale = intervallo[0:intervallo.find("-")]
            giorno_finale = intervallo[intervallo.find("-") + 1:]
            if giorni_della_settimana == "":
                giorni_della_settimana = elenco(giorno_iniziale, giorno_finale)
            else:
                giorni_della_settimana = giorni_della_settimana + ";" + elenco(giorno_iniziale, giorno_finale)
    return giorni_della_settimana

def elenco(giorno_iniziale, giorno_finale):
    if giorno_iniziale == "L":
        if giorno_finale == "Ma":
            return "L;Ma"
        elif giorno_finale == "Me":
            return "L;Ma;Me"
        elif giorno_finale == "G":
            return "L;Ma;Me;G"
        elif giorno_finale == "V":
            return "L;Ma;Me;G;V"
        elif giorno_finale == "S":
            return "L;Ma;Me;G;V;S"
        elif giorno_finale == "D":
            return "L;Ma;Me;G;V;S;D"
    elif giorno_iniziale == "Ma":
        if giorno_finale == "Me":
            return "Ma;Me"
        elif giorno_finale == "G":
            return "Ma;Me;G"
        elif giorno_finale == "V":
            return "Ma;Me;G;V"
        elif giorno_finale == "S":
            return "Ma;Me;G;V;S"
        elif giorno_finale == "D":
            return "Ma;Me;G;V;S;D"
    elif giorno_iniziale == "Me":
        if giorno_finale == "G":
            return "Me;G"
        elif giorno_finale == "V":
            return "Me;G;V"
        elif giorno_finale == "S":
            return "Me;G;V;S"
        elif giorno_finale == "D":
            return "Me;G;V;S;D"
    elif giorno_iniziale == "G":
        if giorno_finale == "V":
            return "G;V"
        elif giorno_finale == "S":
            return "G;V;S"
        elif giorno_finale == "D":
            return "G;V;S;D"
    elif giorno_iniziale == "V":
        if giorno_finale == "S":
            return "v;S"
        elif giorno_finale == "D":
            return "v;S;D"
    elif giorno_iniziale == "S":
        if giorno_finale == "D":
            return "S;D"
    
dfHRS['giorni_della_settimana'] = dfHRS.apply(lambda x: giorni_della_settimana(x['giorni']), axis = 1)
```

```python
def ora(intervallo):
    ora = intervallo.replace(":", ".")
    ora = ora.replace(".15", ".25")
    ora = ora.replace(".30", ".50")
    ora = ora.replace(".45", ".75")
    return float(ora)

def ore_di_apertura(orario):
    ore_di_apertura = 0
    for intervallo in orario.split(';'):
        ore_di_apertura += (ora(intervallo[intervallo.find("-") + 1:]) - ora(intervallo[0:(intervallo.find("-"))]))
    return ore_di_apertura

dfHRS['ore_di_apertura'] = dfHRS.apply(lambda x: ore_di_apertura(x['orario']), axis = 1)
```

```python
import datetime
from datetime import date

def giorno_della_settimana(giorno):
    if giorno == 0:
        return "L"
    elif giorno == 1:
        return "Ma"
    elif giorno == 2:
        return "Me"
    elif giorno == 3:
        return "G"
    elif giorno == 4:
        return "V"
    elif giorno == 5:
        return "S"
    elif giorno == 6:
        return "D"
    return ""

data_iniziale = date(2021, 1, 1)

data_finale = date(2021, 4, 1)

intervallo = [data_iniziale + datetime.timedelta(days = x) for x in range(0, (data_finale - data_iniziale).days)]

df = pd.DataFrame(columns = ['biblioteca', 'giorno', 'ore_di_apertura'])

for data in intervallo:
    for index, row in dfHRS.iterrows():
        if row["inizio"] <= data and row["fine"] >= data:
            if row["giorni_della_settimana"].find(giorno_della_settimana(data.weekday())) == - 1:
                df = df.append({'biblioteca': row["biblioteca"], "giorno": str(data), "ore_di_apertura": 0}, ignore_index = True)
            else:
                df = df.append({'biblioteca': row["biblioteca"], "giorno": str(data), "ore_di_apertura": row["ore_di_apertura"]}, ignore_index = True) 
                
df.sort_values(by = ["biblioteca", "giorno"], inplace = True)
```

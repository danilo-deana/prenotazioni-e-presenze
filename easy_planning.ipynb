# Anno, mese, giorno, settimana, intervallo, ore

```python
def giorno(prenotazione):
    prenotazione = prenotazione[prenotazione.find(" ") + 1:]
    prenotazione = prenotazione.strip()
    prenotazione = prenotazione[0:prenotazione.find(" ")]
    return prenotazione[6:] + "-" + prenotazione[3:5] + "-" + prenotazione[0:2]

def mese(giorno):
    return int(giorno[5:7])

def anno(giorno):
    return int(giorno[0:4])

def settimana(anno, mese, giorno):
    import datetime
    return datetime.date(anno, mese, int(giorno[8:])).isocalendar()[1]

def intervallo(prenotazione):
    prenotazione = prenotazione[prenotazione.find(" ") + 1:]
    prenotazione = prenotazione.strip()
    return prenotazione[prenotazione.find(" ") + 1:]

def ore(intervallo):
    ora_iniziale = intervallo[0:intervallo.find("-")]
    ora_iniziale = ora_iniziale.replace(":", ".")
    ora_iniziale = ora_iniziale.replace(".30", ".50")
    ora_finale = intervallo[intervallo.find("-") + 1:]
    ora_finale = ora_finale.replace(":", ".")
    ora_finale = ora_finale.replace(".30", ".50")
    return float(ora_finale) - float(ora_iniziale)

df['giorno'] = df.apply(lambda x: giorno(x['Periodo prenotazione']), axis = 1)

df['mese'] = df.apply(lambda x: mese(x['giorno']), axis = 1)

df['anno'] = df.apply(lambda x: anno(x['giorno']), axis = 1)

df['settimana'] = df.apply(lambda x: settimana(x['anno'], x['mese'], x['giorno']),
    axis = 1)

df['intervallo'] = df.apply(lambda x: intervallo(x['Periodo prenotazione']), axis = 1)

df['ore'] = df.apply(lambda x: ore(x['intervallo']), axis = 1)
```

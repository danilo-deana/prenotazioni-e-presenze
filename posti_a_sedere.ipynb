# Posti a sedere

```python
settimana = 41
```

```python
import pandas as pd

dfSTS = pd.read_csv('csv' + "/" + "posti_a_sedere.csv", sep = "\t")
```

```python
s = dfSTS[dfSTS['Settimana'] <= settimana].groupby(['Settimana'])['Posti'].sum()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize = (11.7, 8.3), dpi = 300)
sns.set_style("ticks")
sns.lineplot(x = s.index, y = s.values)
sns.despine(right = True)
plt.grid(b = None, which = "both", axis = "y")
plt.xlabel("Settimana dell'anno")
plt.ylabel("Posti a sedere")
plt.show()
```

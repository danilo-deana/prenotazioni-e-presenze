# Orraio di apertura

```python
settimana = 41
```

```python
import pandas as pd

dfORR = pd.read_csv('csv' + "/" + "orari_di_apertura.csv", sep = "\t")
```

```python
def durata(inizio, fine):
    return fine-inizio

dfORR['durata'] = dfORR.apply(lambda x: durata(x['inizio'], x['fine']), axis = 1)
```

```python
dfORR.info()
```

```python
s = dfORR[(dfORR['Settimana'] <= settimana) & (dfORR['durata'] != 0)].groupby(['Settimana'])['durata'].mean()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize = (11.7, 8.3), dpi = 300)
sns.set_style("ticks")
sns.lineplot(x = s.index, y = s.values)
sns.despine(right = True)
plt.grid(b = None, which = "both", axis = "y")
plt.xlabel("Settimana dell'anno")
plt.ylabel("Ore di aperture giornaliere (media)")
plt.show()
```

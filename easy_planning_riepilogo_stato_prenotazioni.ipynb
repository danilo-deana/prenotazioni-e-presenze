# Variabili

```python
from datetime import datetime

anno = datetime.now().year

settimana = 40
```

# Diagramma

```python
import pandas as pd

dfLBR = pd.read_csv("csv" + "/" + "biblioteche.csv", sep = "\t")
```

```python
import pandas as pd

df = pd.read_csv("csv" + "/" + "easy_planning.csv", sep = "\t")
```

```python
s = df.groupby(['anno', 'settimana', 'biblioteca', 'intervallo', 'Stato gestione']).size()

dfTMP = s.to_frame(name = 'prenotazioni')

dfTMP.reset_index(inplace = True)
```

```python
from IPython.display import clear_output
import matplotlib.pyplot as plt
import seaborn as sns

for sigla in dfLBR['sigla']:
    if sigla != "San Paolo":
        plt.figure(figsize = (11.7, 8.3), dpi = 300)
        sns.set_style("ticks")
        sns.barplot(data = dfTMP[(dfTMP["anno"] == anno) & (dfTMP["settimana"] == settimana) &
            (dfTMP["biblioteca"] == sigla)],
            x = "intervallo", y = "prenotazioni" , hue = "Stato gestione",
            hue_order = ['evaso', 'non evaso', 'in attesa'])
        sns.despine(right = True)
        plt.grid(b = None, which = "both", axis = "y")
        plt.xlabel("")
        plt.ylabel("Prenotazioni")
        plt.xticks(rotation = 90)
    
        if sigla == 'Medicina':
            plt.savefig("immagini" + "/" + "Easy_Planning_stato prenotazioni_Polo centrale_anno" + str(anno) + '_settimana_' + str(settimana) + ".jpg",
                dpi = 300)
        else:
            plt.savefig("immagini" + "/" + "Easy_Planning_stato prenotazioni_anno_" + sigla + '_' + str(anno) + '_settimana_' + str(settimana) + ".jpg",
                dpi = 300)
        plt.show()
        clear_output(wait = True)
```
